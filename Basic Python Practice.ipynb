# Problem_7: Write a program that finds the maximum of three numbers.

```python
number_1=int(input("Enter First number:"))
number_2=int(input("Enter Second number:"))
number_3=int(input("Enter Third number:"))

if (number_1>=number_2) & (number_1>=number_3):
    print("Maximum number is:", number_1)

if (number_2>number_1) & (number_2>number_3):
    print("Maximum number is:", number_2)

if (number_3>number_1) & (number_3>number_2):
    print("Maximum number is:", number_3)
```

# Problem_14: Write a program that checks if a given number is prime or not.

```python
from IPython.display import Image
Image(r"C:\Users\Lenovo\Desktop\images\7.jpg")
```

```python
number=int(input("Enter the number:"))
counter=0
if (number==1):
    print(number,"is prime number")
else:
    for i in range(2,number):
        if(number%i==0):
            counter=1

if counter==0:
    print(number,"is prime number")
else:
    print(number,"is not prime number")
```

# Problem_17: Write a program that reverses a given number.

```python
number=int(input("Enter the number:"))
reverse_number=0

while(number!=0):
    remainder = number%10
    reverse_number =reverse_number*10+remainder
    number=number//10
    
print("Reverse number is:",reverse_number)
```

# Problem_19: Write a program that generates a random number and allows the user to guess it.

```python
import random
n = random.randrange(1,10)
guess = int(input("Enter any number: "))

while n!= guess:
    if guess < n:
        print("Too low")
        guess = int(input("Enter number again: "))
    elif guess > n:
        print("Too high")
        guess = int(input("Enter number again: "))
    else:
      break
print("Your guess is right!!")
```
